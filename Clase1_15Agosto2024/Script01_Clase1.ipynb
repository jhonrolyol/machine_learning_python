# Uso de numpy con matplotlib

```python
# Grafico 1
# El objeto primaria para construir/obtener un grafico es una lista

# Definamos los ejes coordenados (2D)
  # Eje-X representa a la variable independiente (puede ser opcional)
  # Eje-Y representa a la variable dependiente (obligatoria)

# Definamos los objetos que van a conformar los ejes
x = [0.5, 1.9, 2.8, 3.2, 7.8]
y = [2, 3.14, 8, 6.9,9.1]

# Diagrama de dispersion
plt.scatter(x, y,s = 50,c = "#ef6a30")

# Fin de las caracteristicas de mi grafico
plt.show()
```

```python
# Diagrama de lineas
plt.plot(x, y)

# Fin
plt.show()
```

```python
# Definamos el dominio de una funcion : linspace
help(np.linspace)
```

```python
# Definamos una funcion  matematica a partir de un dominio y de una regla de correspondencia

# Dominio
dom1 = np.linspace(0,66, 200)

# Regla de correspondencia  : Definamos una funcion matematica
y1 = np.sin(dom1**2)*np.exp(dom1**(0.3))

# Podemos definir esta regla de correspondencia usando una funcion
def f1(x):
  return np.sin(x**2)*np.exp(x**(0.3))

# Definamos otra vez a los elementos de y1
y2 = f1(dom1)

# Tambien podemos definir la regla de correspondencia de la nuesta funcion matematica
# usando una funcion lambda
f2 = lambda x: np.sin(x**2)*np.exp(x**(0.3))

# Definamos otra vez los elementos de y1
y3 = f2(dom1)

# Mostremos y1 , y2 e y3 en diferentes ventanas

#
plt.figure(num = 1)
plt.plot(dom1, y1)

#
plt.figure(num =2 )
plt.plot(dom1, y2)

#
plt.figure(num = 3)
plt.plot(dom1,y3)

# Fin
plt.show()
```

```python
# Grafico3
# Coloquemos en una misma ventana dos conjuntos de pares ordenados

# Dominio : La configuracion del dominio define el conjunto de puntos sobre los cuales
# se construye la variable dependiente
x = np.linspace(1, 2, 20)

# Definamos dos reglas de correspondencia : dos variables dependientes
y1 = np.sin((x+3.14)/(x**(3.14)))
y2 = x**(3.14)/(x*3.14)

# Estos dos conjuntos de pares ordenados : (x,y1) (x,y2)
# Los voy a graficar en una misma ventana
plt.plot(x,y1, "r+",x,y2, "g--")

# Fin
plt.show()
```

```python
# Grafico 4
# Consideremos dominio diferentes para cada regla de correspondencia

#
x1 = np.linspace(2,8, 100)
y1 = np.sin((x1+3.14)/(x1**(3.14)))

#
x2 = np.linspace(3,7, 40)
y2 = x2**(3.14)/(x2*3.14)

# Remarquemos algunos pares ordenados de (x2,y2) : COnsideremos x2 = 4,5,6
x2_prima = np.array([4,5,6])
y2_prima = x2_prima**(3.14)/(x2_prima*3.14)

# Mostremos estos pares ordenados
plt.plot(x1,y1, "b--", x2,y2, "g*", x2_prima, y2_prima, "k*")

# Fin
plt.show()
```

```python
# Grafico 5 : Agregar mas espacio en blanco a los ejes del grafico + grafico4
#
x1 = np.linspace(2,8, 100)
y1 = np.sin((x1+3.14)/(x1**(3.14)))

#
x2 = np.linspace(3,7, 40)
y2 = x2**(3.14)/(x2*3.14)

# Remarquemos algunos pares ordenados de (x2,y2) : COnsideremos x2 = 4,5,6
x2_prima = np.array([4,5,6])
y2_prima = x2_prima**(3.14)/(x2_prima*3.14)

# Mostremos estos pares ordenados
plt.plot(x1,y1, "b--", x2,y2, "g*", x2_prima, y2_prima, "k*")

# Modifiquemos los limites de los ejes cartesianos
plt.xlim(1,9)
plt.ylim((-5,25))

# Fin
plt.show()
```

```python
# Grafico 6: Grafico5 + etiquetas
#
x1 = np.linspace(2,8, 100)
y1 = np.sin((x1+3.14)/(x1**(3.14)))

#
x2 = np.linspace(3,7, 40)
y2 = x2**(3.14)/(x2*3.14)

# Remarquemos algunos pares ordenados de (x2,y2) : COnsideremos x2 = 4,5,6
x2_prima = np.array([4,5,6])
y2_prima = x2_prima**(3.14)/(x2_prima*3.14)

# Mostremos estos pares ordenados
plt.plot(x1,y1, "b--", x2,y2, "g*", x2_prima, y2_prima, "k*")

# Modifiquemos los limites de los ejes cartesianos
plt.xlim(1,9)
plt.ylim((-5,25))

# Agreguemos etiquetas
plt.xlabel("Variable Independiente")
plt.ylabel("Variable Dependiente")
plt.title("Titulo del grafico")


# Fin
plt.show()
```

```python
# Grafico 7 : grafico 6 + personalizacion de las fuentes en las etiquetas
#
x1 = np.linspace(2,8, 100)
y1 = np.sin((x1+3.14)/(x1**(3.14)))

#
x2 = np.linspace(3,7, 40)
y2 = x2**(3.14)/(x2*3.14)

# Remarquemos algunos pares ordenados de (x2,y2) : COnsideremos x2 = 4,5,6
x2_prima = np.array([4,5,6])
y2_prima = x2_prima**(3.14)/(x2_prima*3.14)

# Mostremos estos pares ordenados
plt.plot(x1,y1, "b--", x2,y2, "g*", x2_prima, y2_prima, "k*")

# Modifiquemos los limites de los ejes cartesianos
plt.xlim(1,9)
plt.ylim((-5,25))

# Agreguemos etiquetas modificando las caracteristicas de la fuente
DictEtiquetas_Ejes = {"fontsize" : 16,
                      "color" : "green",
                      "fontstyle": "italic"}

DictEtiquetas_Titulo = {"color": "orange",
                        "fontsize": 22,
                        "fontstyle": "oblique",
                        "fontweight": "bold",
                        "alpha" : 0.5 }
plt.xlabel("Variable Independiente", DictEtiquetas_Ejes)
plt.ylabel("Variable Dependiente", DictEtiquetas_Ejes)
plt.title("Titulo del grafico", DictEtiquetas_Titulo)


# Fin
plt.show()
```

## Analisis Bivariado

```python
# variable1 : LocationAbbr
# variable2 : Data_Value

data.loc[:, ["LocationAbbr", "Data_Value"]]
```

```python
# Forma de calcular el minimo valor de Data_Value para el estado "PA"
data.loc[data.LocationAbbr == "PA", "Data_Value"].min()
```

```python
# Diccionario donde almacenaremos el estado y el valor minimo de Data_Value para
# cada estado en particular

Dict_DataValue = {}.fromkeys(data.LocationAbbr.unique())

for estado in data.LocationAbbr.unique():
  Dict_DataValue[estado] = data.loc[data.LocationAbbr == estado, "Data_Value"].min()

pd.DataFrame.from_dict(Dict_DataValue,orient = "index",columns= ["Min_DataValue"]).sort_values(by = "Min_DataValue", ascending = False).plot(kind = "bar",
                                                                                                                                             figsize = (12,12))
```
