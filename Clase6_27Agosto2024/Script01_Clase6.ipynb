```python
# MOdulos
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset propio de sklearn : iris
from sklearn import datasets

# Aprendizaje no supervizado
from sklearn.cluster import KMeans

# Dataset : iris
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2]
y = iris.target
X
```

```python
# Diagrama de dispersion
plt.scatter(X[:,0],y = X[:,1], c = y)
```

```python
# Instanciamos la clase a ajustar
kmean1 = KMeans(n_clusters = 3)
kmean1.fit(X)
```

```python
# OBservemos el modelo
kmean1
```

```python
# Obtencion de los centroides
kmean1.cluster_centers_
```

```python
# Comparemos los resultados obtenidos con los datos iniciales

NuevasEtiquetas = kmean1.labels_

fig, axes = plt.subplots(1,2, figsize = (15,5))
axes[0].scatter(X[:,0], X[:,1], c = y)
axes[1].scatter(X[:,0], X[:,1], c = NuevasEtiquetas)
axes[0].set_title("Data Original")
axes[1].set_title("Prediccion de KMeans")
plt.show()
```
