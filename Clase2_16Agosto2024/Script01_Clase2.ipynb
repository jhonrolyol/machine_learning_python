```python
# Agreguemos a este pequeño analisis algo de informacion temporal
sudamerica = globalT.loc[globalT.region_txt == "South America", ["iyear", "country_txt"]]
sudamerica

# Para los incidentes ocurridos en sudamerica , tenemos informacion temporal (iyear)
# y mas informacion espacial en la columna country_txt

```

```python
# Para la informacion obtenida en la celda anterior, podemos observar la distribucion
# de incidentes en la lineal del tiempo (iyear) para un pais sudamericano en particular
sudamerica.loc[sudamerica.country_txt == "Peru", "iyear"].value_counts().sort_index().plot(kind = "bar",
                                                                                           figsize = (10,10))

# Carguemos matplotlib
import matplotlib.pyplot as plt
plt.xlabel("Años")
plt.ylabel("Numero de ataques terroristas")
plt.title("Pais : PERÚ")
# Subtitulo
plt.suptitle("Base datos : https://www.kaggle.com/datasets/START-UMD/gtd")

# Final
# plt.show()

# Guardemos la imagen en disco duro
plt.savefig("Incidentes_PERU.png")
```

```python
# Generemos el mismo grafico para todos los paises de sudamerica
for pais in sudamerica.country_txt.unique():
  sudamerica.loc[sudamerica.country_txt == pais, "iyear"].value_counts().sort_index().plot(kind = "bar",
                                                                                           figsize = (10,10))

  plt.xlabel("Años")
  plt.ylabel("Numero de ataques terroristas")
  plt.title("Pais : "  + pais)
  # Subtitulo
  plt.suptitle("Base datos : https://www.kaggle.com/datasets/START-UMD/gtd")

  # Final
  # plt.show()

  # Guardemos la imagen en disco duro
  plt.savefig("Incidentes_" + pais + ".png")
```

# Analisis Bivariado

```python
# Consideremos analizar un par de variables de diferente naturaleza , manteniendo la
# informacion temporal.
# Columna Categorica : region_txt
# Columnas Cuantitavica : nkill

globalT.loc[: , ["iyear", "region_txt", "nkill"]]

# Podemos observar que en la columna nkill hay valores NaN (valores faltantes)
```

```python
# Mostremos los valores faltantes para estas 3 columnas : "iyear", "region_txt" y "nkill"
globalT.loc[: , ["iyear", "region_txt", "nkill"]].isnull().sum()
```

```python
# Consideremos rellenar los valores faltantes de la columna nkill
# con el valor de 0
globalT.loc[: , ["iyear", "region_txt", "nkill"]].fillna(value = 0.0)
```

```python
# Calculemos (para cada una de las regiones) el total de victimas
globalT.loc[globalT.region_txt == "Central Asia", "nkill"].sum()
```

```python
# Repetir los calculos de la celda anterior para todas las regiones que se
# encuentran en la columna region_txt

# Construir un diccionario :
  # llave : region
  # valor : total de victimas

# Construyamos una lista con los elementos que componen a la columnas region_txt
ListaRegiones = globalT.region_txt.unique()

# Diccionar con llaves en ListaRegiones y valores seteados en None
DictRegiones_nkill = {}.fromkeys(ListaRegiones)

# Usaremos una estructura de repeticion para rellenar los valores del
# diccionario DictRegiones_nkill
for reg in ListaRegiones:
  DictRegiones_nkill[reg] = globalT.loc[globalT.region_txt == reg, "nkill"].sum()

DictRegiones_nkill
```

```python
# Transformemos el diccionario DictRegiones_nkill a objeto de tipo dataframe
# para luego mostrar un diagrama de barras
pd.DataFrame.from_dict(DictRegiones_nkill, orient = "index", columns = ["NumFallecidos"]).sort_values(by = "NumFallecidos").plot(kind = "bar")
```

```python
# Tarea :
  # Otra estrategia : Eliminar las observaciones que tienen valores faltantes en la columna nkill
  # y repetir el grafico final

```

```python
# Para el siguiente analisis consideremos la informacion temporal
tmp1 = globalT.loc[:,["iyear", "region_txt", "nkill"]]

# Para esta region en particular , deseo obtener como evoluciona el numero de fallecidos (nkill)
# por año
tmp1.loc[tmp1.region_txt == "Central Asia", "iyear" ].value_counts().sort_index().plot()
plt.title("Region : Central Asia")
plt.xlabel("Años")
plt.ylabel("Numero de Fallecidos")

# FIN
plt.show()
```

```python
# Para almacenar las imagenes en disco dura, hay que tener cuidado de que
# los nombres de los archivos no posean espacios en blanco
for region in ListaRegiones :
  print(region.replace(" ", "_"))
```

```python
# Repliquemos el grafico obtenido en la celda anterior para todas las regiones

for region in ListaRegiones:
  tmp1.loc[tmp1.region_txt == region, "iyear" ].value_counts().sort_index().plot()
  plt.title("Region " + region)
  plt.xlabel("Años")
  plt.ylabel("Numero de Fallecidos")

  # Almacenemos la imagen/figura a disco duro
  plt.savefig("NumFallecidos_"+ region.replace(" ", "_") + ".png")
  # FIN
  plt.show()

```
